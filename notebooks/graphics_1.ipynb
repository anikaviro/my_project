## 1. What are the top 3 areas in plastic waste generation?

In order to answer this question, we are going to plot the evolution of all areas with their evolution in the 20 years measured. We will use matplotlib and seaborn to generate the graphics.

```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
```

```python
data = pd.read_csv("../data/clean/totalbyyearandcountry.csv")
data.head()
```

```python
ax = sns.lineplot(data, x="year", y="m_tons_plastic", hue="location");
sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))
sns.set(rc={'figure.figsize':(10,10)})
ax.set_ylabel("Plastic waste (in millions of tons)")
ax.set_title("Evolution of plastic waste generation")
ax.set_xticks(range(2000,2019,2));
```

With this, we can easily see that for the last 20 years, United States, China and the OECD countries of the European Union are the top 3 areas in terms of generation of plastic waste. It is interesting to see that China surpassed OECD EU in 2011.
